# Model training

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./data/gemstone.csv")
```

```python
data.drop(labels="id",axis=1,inplace=True)
```

```python
data.head()
```

```python
X=data.drop(labels=["price"],axis=1)
```

```python
Y=data[["price"]]
```

```python
X
```

```python
Y
```

```python
categorical_column = X.select_dtypes(include="object").columns
```

```python
numerical_column = X.select_dtypes(exclude="object").columns
```

```python
# define the custom ranking for each ordinal variables

cut_categories = ["Fair","Good","Very Good","Premium","Ideal"]
clarity_categories = ["I1","SI2","SI1","VS2","VS1","VVS2","VVS1","IF"]
color_categories = ["D","E","F","G","H","I","J"]
```

```python
from sklearn.impute import SimpleImputer ## Handling the missing values
from sklearn.preprocessing import StandardScaler # Handling Feature Scaling
from sklearn.preprocessing import OrdinalEncoder # Ordinal encoding

# pipeline for encoding
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
num_pipeline = Pipeline(
    
    
    steps=[
        
        ('imputer', SimpleImputer()),
        ('scaler', StandardScaler())
        
    ]
    
     
    
)
```

```python
cat_pipeline = Pipeline(
    
    
    steps=[
        
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('ordinalencoder', OrdinalEncoder(categories=[cut_categories,color_categories,clarity_categories]))
        
    ]
    
     
    
)
```

```python
preprocessor = ColumnTransformer(
    
    [
        ('num_pipeline', num_pipeline,numerical_column),
        ('cat_pipeline', cat_pipeline,categorical_column)
        
    ]
    
)
```

```python
# train test split

from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.30,random_state=30)
```

```python
preprocessor.fit_transform(X_train)
```

```python
preprocessor.transform(X_test)
```
